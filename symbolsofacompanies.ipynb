```python
import requests

def get_all_symbols(api_key):
    url = f'https://www.alphavantage.co/query?function=LISTING_STATUS&apikey={api_key}'
    response = requests.get(url)

    # Check if the response status code is OK (200)
    if response.status_code != 200:
        print(f"Error: Received status code {response.status_code}")
        print(response.text)
        return []

    try:
        data = response.json()
    except requests.exceptions.JSONDecodeError:
        print("Error: Unable to parse JSON response")
        print(response.text)
        return []

    symbols = [item['symbol'] for item in data]
    return symbols

api_key = '1HRCG3M2TLBM8PIS'
symbols = get_all_symbols(api_key)
print(symbols)
```

```python
import requests
import csv
from io import StringIO

def get_all_symbols(api_key):
    url = f'https://www.alphavantage.co/query?function=LISTING_STATUS&apikey={api_key}'
    response = requests.get(url)

    # Check if the response status code is OK (200)
    if response.status_code != 200:
        print(f"Error: Received status code {response.status_code}")
        print(response.text)
        return []

    try:
        # Read the response as CSV
        csv_content = response.text
        csv_reader = csv.DictReader(StringIO(csv_content))

        # Extract symbols from the CSV data
        symbols = [row['symbol'] for row in csv_reader]
        return symbols

    except Exception as e:
        print("Error: Unable to parse CSV response")
        print(str(e))
        print(response.text)
        return []

api_key = '1HRCG3M2TLBM8PIS'
symbols = get_all_symbols(api_key)
print(symbols)
```

```python
import requests
import csv
from io import StringIO

def get_all_symbols_and_names(api_key):
    url = f'https://www.alphavantage.co/query?function=LISTING_STATUS&apikey={api_key}'
    response = requests.get(url)

    # Check if the response status code is OK (200)
    if response.status_code != 200:
        print(f"Error: Received status code {response.status_code}")
        print(response.text)
        return []

    try:
        # Read the response as CSV
        csv_content = response.text
        csv_reader = csv.DictReader(StringIO(csv_content))

        # Extract symbols and names from the CSV data
        symbols_and_names = [{'symbol': row['symbol'], 'name': row['name']} for row in csv_reader]
        return symbols_and_names

    except Exception as e:
        print("Error: Unable to parse CSV response")
        print(str(e))
        print(response.text)
        return []

api_key = '1HRCG3M2TLBM8PIS'
symbols_and_names = get_all_symbols_and_names(api_key)
print(symbols_and_names)
```
